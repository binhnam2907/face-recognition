```python
def pipline(imagepath1, imagepath2):
    faces1 = extract_faces(img_path = imagepath1, align = True)
    faces2 = extract_faces(img_path = imagepath2, align = True)


    faces1 = faces1[0]
    faces2 = faces2[0]
    print(faces1.shape, faces2.shape)
    feature1, feature2 = return_feature(faces1), return_feature(faces2)
    print(feature1.shape, feature2.shape)
    #loss_arr = []

    #loss_arr.append(arcface_loss(feature1, feature2))
    
    return arcface_loss(feature1, feature2)
```

```python
count = 0 
for i in range(1, 20):
    path = "data/sample" + str(i) + ".jpg"
    loss = pipline("data/sample0.jpg", path)
    if loss < 0.1:
        print("True", loss)
    else:
        print("False: " ,  loss)
        count += 1
```
